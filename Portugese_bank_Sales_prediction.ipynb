### In this following dataset, we have 21 columns for predicting bank product sales.

   **Age:**
       
        In this column they have given different age of customers.
   
   **Job:**
        
        In this column they have given different job roles of customers like housemaid, services,admin,services,technician,
        blue-collar etc.
        
   **Martial:**
        
        In martial column there are different categories which are 'divorced','married','single','unknown'.
        
   **Education:**
   
        In education field they have provided different education done by the customers like 'basic.4y','basic.6y',basic.9y'
        'high.school','illiterate','professional.course','university.degree'and'unknown'.
        
   **Default:**

        In default field they have gave customers who had credit or not.
      
   **Housing:**

        In housing column bank has provided the customer data who had taken housing loan or not.
        
   **Loan:**

        In loan field, the data shows that customer had taken personal loan or not.

###### related with the last contact of the current campaign:  
   
   **Contact:**

        In contact column the bank have provided the communication type as 'cellular' and 'telephone'.
        
   **Month:**

        The given "month column" represents the last contact month, indicating the month in which customers receive calls.
        
   **Day_of_week:**

         The "day_of_week" column represents the day of the week on which customers are receiving calls.
        
   **Duration:**

         Duration shows how many seconds that bank interacts with the customers. 

##### other attributes:        
   
   **Campaign:**

         Number of contacts performed during this campaign and for this client numeric, includes last contact.
          
   **Pdays:**

         Number of days that passed by after the client was last contacted from a previous campaign numeric=999 means client             was not previously contacted.
          
   **Previous:**

         Number of contacts performed before this campaign and for this client.
   
   **Poutcome:**

         Poutcome represents the previous marketing campaign like 'failure','nonexistant','success'.
         
#####  social and economic context attributes
  
   **emp.var.rate:**

         In banking context, it might be used to assess how changes in the employment situation could impact the bank's                  operations, customer behavior, or credit risk.

   **cons.price.idx:**

          The consumer price index is widely used economic indicator that helps assess inflation and its impact on consumers.
         
   **cons.conf.idx:**

         In a banking context, the "cons.conf.idx" could be used to understand how consumer confidence levels might affect                various banking activities, such as borrowing, spending, or investment decisions.
       
   **euribor3m:**
         
         It is one of the most important benchmark interest rates in the eurozone and is widely used in various financial                transactions, including loans, mortgages, and derivative contracts.
         
   **nr.employes:**

          "nr.employed" typically refers to the number of employees, specifically within the banking or financial institution
          being analyzed.
          
   **Target**
        
        The 'y' considered as the ouput column which we have to predict the values by using the above independent features.

## 7.MODEL CREATION

### Spliting X and y

```python
X=data.drop('y',axis=1)
y=data['y']
```

```python
#spliting train and test data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.30)
```

```python
X_train.value_counts().sum()
```

```python
X_test.value_counts().sum()
```

### Balancing the Data

```python
data.y.value_counts()
```

* datas are imbalanced so we can use smote to balance the data

```python
from imblearn.over_sampling import SMOTE
smote=SMOTE()
```

```python
X_smote,y_smote=smote.fit_resample(X_train,y_train)
```

```python
# for comparing actual and smote values we use collections 
from collections import Counter
print("Actual Class:",Counter(y_train))
print("SMOTE Class:",Counter(y_smote))
```

* Thus we have balanced the data now we import different algorithms and perform predictions

## Modeling with Logistic Regression


```python
#importing the model
from sklearn.linear_model import LogisticRegression
lg=LogisticRegression()
```

```python
#training the model
lg.fit(X_train,y_train)
```

```python
#predict the model
y_pred=lg.predict(X_test)
y_train_pred=lg.predict(X_train)
```

```python
y_pred
```

### Model Evaluation for Logistic Regression

### Testing data

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, precision_score,recall_score,f1_score
```

```python
lr_accuracy=accuracy_score(y_test,y_pred)
```

```python
lr_accuracy
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
fpr, tpr, thresholds = roc_curve(y_test,y_pred)
roc_auc = roc_auc_score(y_test,y_pred)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = {:.2f})'.format(roc_auc))
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```

### Insights 
      For Logistic Regression we have got '91%'of accuracy. so, our model has predicted 9% of which customer will buy the product
